```python
import numpy as np
import os
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
```

```python
employee_df = pd.read_csv('employee.csv')
employee_df.set_index('employee_id', drop = True, inplace = True)

employee_df.head(5)
```

```python
engine = create_engine('sqlite:///hr.sqlite', echo=True)
sqlite_connection = engine.connect()
```

```python
sqlite_table = "employee"
employee_df.to_sql(sqlite_table, sqlite_connection, if_exists='replace')
```

```python
sqlite_connection.close()
```

```python
# Create a SQL connection to our SQLite database
con = sqlite3.connect("hr.sqlite")

cur = con.cursor()

# The result of a "cursor.execute" can be iterated over by row
for row in cur.execute('SELECT * FROM employee;'):
    print(row)
```

```python
con.close()
```
